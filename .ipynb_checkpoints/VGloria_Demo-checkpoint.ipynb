# Initialize model

```python
import numpy as np
from torch import nn
import torch
import yaml
import os
import torch
from torchvision import transforms
from torch.utils.data import DataLoader
from torchvision.datasets.coco import CocoCaptions
import json
from tqdm import tqdm

from PIL import Image
import matplotlib.pyplot as plt
import sys

sys.path.append('/user/home/a.simplicio/VGloria')
from fromage import models
from fromage import utils
#from sklearn.metrics import recall_score
```

```python
model = models.Fromage.from_pretrained("NOVA-vision-language/V-GlorIA")
```

```python
#load model
model_dir = '/user/home/a.simplicio/VGloria/VGlorIA_model'
model = models.load_fromage(model_dir)
```

```python
def get_pixel_values_from_path(path: str, feature_extractor):
    """Helper function for getting images pixels from a local path."""
    img = Image.open(path)
    img = img.resize((224, 224))
    img = img.convert('RGB')
    pixel_values = utils.get_pixel_values_for_model(feature_extractor, img)
    if torch.cuda.is_available():
        pixel_values = pixel_values.bfloat16()
        pixel_values = pixel_values.cuda()
    return pixel_values[None, ...]
```

```python
# Create lists to save the images from MS-COCO
img_files = []
all_visual_embs = []
```

```python
for filename in os.listdir('/storagebk/datasets/ms-coco/val2017'):
    img_files.append(filename)
    images = get_pixel_values_from_path(
            os.path.join('/storagebk/datasets/ms-coco/val2017', f'{filename}'),
            model.model.feature_extractor)
    visual_embs = model.model.get_visual_embs(images, mode='retrieval')
    all_visual_embs.append(visual_embs.cpu().detach().float().numpy())
```

```python
all_text_embs = []
for caption in ["Uma girafa [RET]"]:
    input_ids = model.model.tokenizer(caption, add_special_tokens=True, return_tensors="pt").input_ids
    input_ids = input_ids.cuda()
    input_embs = model.model.input_embeddings(input_ids)  # (N, T, D)
    generated_ids, output_embs, _ = model(input_embs, None, None, generate=True, num_words=1, temperature=0.0)
    embeddings = output_embs[0]
    ret_emb = embeddings[:, -1, :]  
    all_text_embs.append(ret_emb.cpu().detach().float().numpy())

```

# Retrieval

```python
# Simple retrieval function caption --> image
def retrieve(caption, topk = 1):
    caption = caption + '[RET]'
    input_ids = model.model.tokenizer(caption, add_special_tokens=True, return_tensors="pt").input_ids
    input_ids = input_ids.cuda()
    input_embs = model.model.input_embeddings(input_ids)  # (N, T, D)
    generated_ids, output_embs, _ = model(input_embs, None, None, generate=True, num_words=1, temperature=0.0)
    embeddings = output_embs[0]
    ret_emb = embeddings[:, -1, :]  
    all_text_embs = [ret_emb.cpu().detach().float().numpy()]

    scores = np.concatenate(all_visual_embs, axis=0)[:, 0, :] @ np.concatenate(all_text_embs, axis=0).T
    scores = torch.tensor(scores).float().T
    _, preds = scores.topk(topk)
    for pred in preds[0]:    
        display(Image.open(f'/storagebk/datasets/ms-coco/val2017/{img_files[pred]}'))
```

```python
retrieve('Um grande relógio')
```

```python
retrieve('Um homem senta-se numa sanita partida enquanto as pessoas passam.')
```
